```python
class User:
    def __init__(self, name, email, address, phone_number, password):
        self.name = name
        self.email = email
        self.address = address
        self.phone_number = phone_number
        self.password = password
        self.order_history = []

class Item:
    def __init__(self, name, description, price):
        self.name = name
        self.description = description
        self.price = price

class Menu:
    def __init__(self, items):
        self.items = items

class Order:
    def __init__(self, user, items, total):
        self.user = user
        self.items = items
        self.total = total

class Restaurant:
    def __init__(self, name, address, menu):
        self.name = name
        self.address = address
        self.menu = menu

    def display_menu(self):
        for item in self.menu.items:
            print(f"{item.name}: {item.description} - {item.price}")

    def place_order(self, user, item_names):
        items = []
        total = 0
        for name in item_names:
            for item in self.menu.items:
                if item.name == name:
                    items.append(item)
                    total += item.price
                    break
        order = Order(user, items, total)
        user.order_history.append(order)
        print(f"Order placed for {user.name}. Total: {total}")

    def display_order_history(self, user):
        print(f"Order history for {user.name}:")
        for order in user.order_history:
            print(f"Items: {[item.name for item in order.items]}. Total: {order.total}")

class Admin:
    def __init__(self, username, password):
        self.username = username
        self.password = password

    def add_item(self, item, menu):
        menu.items.append(item)
        print(f"Added {item.name} to the menu")

    def remove_item(self, item, menu):
        menu.items.remove(item)
        print(f"Removed {item.name} from the menu")

items = [
    Item("Tandoori Chicken", "A delicious Tandoori Chicken", 240),
    Item("Burger", "A juicy burger with all the fixings", 320),
    Item("Truffle Cake", "this cake is very delicious", 900),
]

menu = Menu(items)

restaurant = Restaurant("Joe's Pizza", "123 Main St.", menu)

users = []

while True:
    choice = input("Would you like to login (2) or register (1)? ")
    if choice == "1":
        name = input("Enter your name: ")
        email = input("Enter your email: ")
        address = input("Enter your address: ")
        phone_number = input("Enter your phone number: ")
        password = input("Enter your password: ")

        user = User(name, email, address, phone_number, password)
        users.append(user)
        print(f"Registration successful! Welcome, {user.name}!")
      
    elif choice == "2":
        email = input("Enter your email: ")
        password = input("Enter your password: ")
        for user in users:
            if user.email == email and user.password == password:
                print(f"Login successful! Welcome, {user.name}!")
                restaurant.display_menu()
                restaurant.place_order(user, ["Pizza", "Fries"])
                restaurant.display_order_history(user)
                break
        else:
            print("Invalid email or password")
  
```
